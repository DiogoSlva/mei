```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import learning_curve
from sklearn.metrics import make_scorer, mean_squared_error
```

# Exerc√≠cio 1.1

```python
df = pd . read_csv ('Grade_Set_1.csv')
print (df)


#df.plot(kind = 'scatter' , X = 'Hours_Studied' , y = 'Test_Grade' , title = 'Grade vs Hours studied')

X = df[['Hours_Studied']] 
y = df['Test_Grade']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

modelo = LinearRegression()
modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)

plt.scatter(X_test, y_test, color='blue', label='Dados reais')
plt.plot(X_test, y_pred, color='red', label='Linha de Regress√£o')
plt.xlabel('Hours of Study')
plt.ylabel('Grade')
plt.title('Regress√£o Linear Univariada')
plt.legend()
plt.show()
```

# Exerc√≠cio 1.2

```python
# Criando o modelo de Regress√£o Linear
modelo = LinearRegression()

# Treinando o modelo com os dados de treino
modelo.fit(X_train, y_train)

# Fazendo previs√µes com o conjunto de teste
y_pred = modelo.predict(X_test)

# Plotando a linha de regress√£o nos dados de treino
plt.scatter(X_test, y_test, color='blue', label='Dados Reais')
plt.plot(X_test, y_pred, color='red', label='Linha de Regress√£o')
plt.xlabel('Horas Estudadas')
plt.ylabel('Nota no Teste')
plt.title('Regress√£o Linear - Nota vs Horas Estudadas')
plt.legend()
plt.show()

# Exibindo os coeficientes da regress√£o
print(f'Coeficiente (inclina√ß√£o): {modelo.coef_[0]}')
print(f'Intercepto: {modelo.intercept_}')

# Avalia√ß√£o do modelo
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Erro quadr√°tico m√©dio (MSE): {mse}')
print(f'Coeficiente de Determina√ß√£o (R¬≤): {r2}')
```
