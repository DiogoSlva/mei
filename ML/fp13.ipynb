# Parte 1

```python
import matplotlib . pyplot as plt
import numpy as np
import pandas as pd

from sklearn.linear_model import Perceptron
from matplotlib . colors import ListedColormap
from sklearn . datasets import make_classification
from sklearn.datasets import load_digits
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import metrics
```

## Perceptron

```python
X, y = make_classification(n_samples=20, n_features=2, n_informative=2, n_redundant=0, weights=[0.5, 0.5], random_state=2017)

colormap = np.array(['r', 'k'])
plt.scatter(X[:, 0], X[:, 1], c=colormap[y], s=40)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
# Load the digits dataset
digits = load_digits()
print('We have %d samples' % len(digits.target))

# Plot the first 64 samples to get a sense of the data
fig = plt.figure(figsize=(8, 8))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.gray_r)
    ax.text(0, 1, str(digits.target[i]), bbox=dict(facecolor='white'))

# Display the plot
plt.show()
```

## Multilayer Perceptrons

```python
X_train , X_test , y_train , y_test = train_test_split(digits.data , digits.target, test_size = 0.2 , random_state = 2017)
print ('Number of samples in training set : % d ' % ( len ( y_train ) ) )
print ('Number of samples in test set : % d ' % ( len ( y_test ) ) )
# Standardise data , and fit only to the training data
scaler = StandardScaler ()
scaler . fit ( X_train )
# Apply the transformations to the data
X_train_scaled = scaler . transform ( X_train )
X_test_scaled = scaler . transform ( X_test )
```

```python
# Initialize ANN classifier ( 1 hidden layer with 100 neurons )
mlp = MLPClassifier ( hidden_layer_sizes = ( 200, 200, 150 ) , activation = 'logistic' ,max_iter = 1000 )
# Train the classifier with the traning data
mlp.fit ( X_train_scaled , y_train )
print ( " Training set score : % f " % mlp . score ( X_train_scaled , y_train ) )
print ( " Test set score : % f " % mlp . score ( X_test_scaled , y_test ) )
# predict results from the test data
X_test_predicted = mlp . predict ( X_test_scaled )
```

# Parte 2

### Exerccio 1

```python
# Load the diabetes dataset
diabetes_df = pd.read_csv('Dataset/Diabetes.csv')
print(diabetes_df.head())

# Selecionar as variveis de interesse
X = diabetes_df[['bmi', 'age']]
y = diabetes_df['class']

# Dividir os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2017)

# Padronizar os dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Inicializar o classificador MLP
mlp = MLPClassifier(hidden_layer_sizes=(100), activation='relu', max_iter=1000, random_state=2017)

# Treinar o classificador com os dados de treinamento
mlp.fit(X_train_scaled, y_train)

# Avaliar o classificador
train_score = mlp.score(X_train_scaled, y_train)
test_score = mlp.score(X_test_scaled, y_test)

print(f"Training set score: {train_score:.6f}")
print(f"Test set score: {test_score:.6f}")
```

```python
# Selecionar as variveis de interesse
X = diabetes_df[['bmi', 'age']]
y = diabetes_df['class']

# Dividir os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2017)

# Padronizar os dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Inicializar o classificador MLP
mlp = MLPClassifier(hidden_layer_sizes=(200,200, 100), activation='logistic', max_iter=1000, random_state=2017)

# Treinar o classificador com os dados de treinamento
mlp.fit(X_train_scaled, y_train)

# Avaliar o classificador
train_score = mlp.score(X_train_scaled, y_train)
test_score = mlp.score(X_test_scaled, y_test)

print(f"Training set score: {train_score:.6f}")
print(f"Test set score: {test_score:.6f}")
```

### Exerccio 3

```python
# Selecionar as variveis de interesse
X = diabetes_df[['preg_count','glucose_concentration','blood_pressure','skin_thickness','serum_insulin','bmi','pedigree_function','age']]
y = diabetes_df['class']

# Dividir os dados em conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2017)

# Padronizar os dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Inicializar o classificador MLP
mlp = MLPClassifier(hidden_layer_sizes=(200,200, 100), activation='logistic', max_iter=1000, random_state=2017)

# Treinar o classificador com os dados de treino
mlp.fit(X_train_scaled, y_train)

# Avaliar o classificador
train_score = mlp.score(X_train_scaled, y_train)
test_score = mlp.score(X_test_scaled, y_test)

print(f"Training set score: {train_score:.6f}")
print(f"Test set score: {test_score:.6f}")
```

### Exerccio 4

```python
# Inicializar o modelo de regresso linear
linear_regressor = LinearRegression()

# Treinar o modelo com os dados de treinamento
linear_regressor.fit(X_train_scaled, y_train)

# Fazer previses nos dados de teste
y_test_pred = linear_regressor.predict(X_test_scaled)

# Avaliar o modelo
mse = mean_squared_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)

print(f"Mean Squared Error: {mse:.6f}")
print(f"R^2 Score: {r2:.6f}")

# Convert continuous predictions to binary class labels
y_train_pred_class = (linear_regressor.predict(X_train_scaled) > 0.5).astype(int)
y_test_pred_class = (y_test_pred > 0.5).astype(int)

print('Train accuracy:', metrics.accuracy_score(y_train, y_train_pred_class))
print('Train confusion matrix\n', metrics.confusion_matrix(y_train, y_train_pred_class))
print('Classification report:\n', metrics.classification_report(y_train, y_train_pred_class))
print('Test accuracy:', metrics.accuracy_score(y_test, y_test_pred_class))
print('Test confusion matrix\n', metrics.confusion_matrix(y_test, y_test_pred_class))
print('Classification report:\n', metrics.classification_report(y_test, y_test_pred_class))
```
